```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import csv

imgs_path = '../images'
sal_inp_path = '../sal_inpaint'
sal_noninp_path = '../sal_noninpaint'
seg_path = '../segmentation'
sod_roi_path = '../sod_mask'

images = os.listdir(sal_inp_path)

for img in images:

    img_name = img.split('_mask.png')[0]+'.jpg'
    sal_inp = Image.open(os.path.join(sal_inp_path,img)).convert('L')
    sal_noninp = Image.open(os.path.join(sal_noninp_path,img_name)).convert('L')
    seg = Image.open(os.path.join(seg_path,img_name)).convert('L')
    sod = Image.open(os.path.join(sod_roi_path,img_name)).convert('L')
        
    sal_inp_np = np.array(sal_inp)
    sal_noninp_np = np.array(sal_noninp)
    seg_np = np.array(seg)

    inp_total = 0
    noninp_total = 0
    sod_inp_total = 0
    sod_noninp_total = 0
    [rows,cols] = sal_inp_np.shape
    #print(img_name)
    for i in range(rows):
        for j in range(cols):

            if seg_np[i,j]==2:
                if sal_inp_np[i,j]!=0:
                    inp_total+=sal_inp_np[i,j]
                if sal_noninp_np[i,j]!=0:
                    noninp_total+=sal_noninp_np[i,j]

            if sod_np[i,j]==255:
                if sal_inp_np[i,j]!=0:
                    sod_inp_total+=sal_inp_np[i,j]
                if sal_noninp_np[i,j]!=0:
                    sod_noninp_total+=sal_noninp_np[i,j]

    if noninp_total==0:
        noninp_total = 1
    roi_increase = (inp_total-noninp_total)/noninp_total
    
    if sod_noninp_total==0:
        sod_noninp_total = 1
    sod_reduce = (sod_inp_total-sod_noninp_total)/sod_noninp_total
    
    saliency_change=[img_name,roi_increase,sod_reduce,noninp_total,inp_total,sod_noninp_total,sod_inp_total]
    with open("saliency_change.csv", "a", newline='',encoding='utf-8') as f:
        writer = csv.writer(f,delimiter=',')
        writer.writerow(saliency_change)
                
```
